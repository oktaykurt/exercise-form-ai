```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('pose-coordinates/lunge_df6.csv')

# # Filter the rows where exercise is 'squat'
# df = df[df['exercise'] == 'squat']

# Parse the coordinates into separate columns
df[['left_shoulder_x', 'left_shoulder_y']] = pd.DataFrame(df['left_shoulder'].apply(ast.literal_eval).to_list(), index=df.index)

# Convert frame_time to datetime
df['frame_time'] = pd.to_datetime(df['frame_time'])

# Set frame_time as index (for time-series plot)
df.set_index('frame_time', inplace=True)

# Plot the x and y coordinates over time
plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
plt.plot(df['left_shoulder_x'])
plt.title('Left Shoulder X over time')
plt.xlabel('Frame time')
plt.ylabel('Left hip X')

plt.subplot(2, 1, 2)
plt.plot(df['left_shoulder_y'])
plt.title('Left Shoulder Y over time')
plt.xlabel('Frame time')
plt.ylabel('Left hip Y')

plt.tight_layout()
plt.show()
```
